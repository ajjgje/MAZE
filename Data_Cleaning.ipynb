# Working with Spotify 2023

```python
# filtering out 2023 years from 2023 csv file

df.columns = df.columns.str.strip()

df['year_published'] = pd.to_numeric(df['year_published'], errors='coerce')

df_2023 = df[df['year_published'] == 2023]

df_2023.head()
```

```python
# deleting cols

df.columns = df.columns.str.strip()
df_2023 = df_2023.drop(columns = ["key","bpm","mode","released_month", "released_day", "artist_count", "danceability_",	"valence_"	,"energy_",	"acousticness_"	,"instrumentalness_"	,"liveness_","speechiness_"])

df_2023.head()
```

# Working with Spotify 2024

```python
# Filtering out 2024 year songs of 2024 csv file
df2['Release Date'] = pd.to_datetime(df2['Release Date'], errors='coerce')

# Filter rows where the year is 2024
df_2024 = df2[df2['Release Date'].dt.year == 2024]

df_2024.head()
```

```python
df_2024 = df_2024.drop(columns=["Explicit Track", "SiriusXM Spins", "TIDAL Popularity"])

df_2024["Release Year"] = df_2024["Release Date"].dt.year
cols = list(df_2024.columns)

col1, col2 = "Release Date", "Release Year" #swapping

i, j = cols.index(col1), cols.index(col2)  # find their positions
cols[i], cols[j] = cols[j], cols[i]        # swap them

df_2024 = df_2024[cols]

df_2024 = df_2024.drop(columns = ["Release Date"])
df_2024.head()
```
